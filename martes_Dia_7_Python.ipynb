##Introducción a la Regresión Lineal

```python
from sklearn.linear_model import LinearRegression
```

#¿Qué es una regresión lineal?
Una regresión intenta predecir un valor numérico a partir de otros datos

Ejemplo:
- Predecir ingresos segun edad
- predecir precio segun tamaño

Aquí usaremos regresión lineal:el tipo más simple

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x = df_clean["edad"], y = df_clean["ingresos"])
plt.xlabel("Edad")
plt.ylabel("Ingresos")
plt.title("Relación entre edad e ingresos")
plt.show()
```

```python
#MATRIZ DE CORRELACION
#df_clean.corr()
#te va a marcar error
#correlación
df_clean[["edad", "ingresos"]].corr()
# 0.0 -> nada
# 0.3 -> debil
# 0.5-0.7 -> moderada
# >0.8 -> fuerte
```

```python
print(df_clean)
```

```python
X = df_clean[["edad"]]
y = df_clean["ingresos"]
modelo = LinearRegression()
modelo.fit(X,y)
```

Tiene relación con la fórmula de la pendiente
que hace fit? Busca los valores de m, y b que mejor se ajustan a tus datos.
Minimiza el error entre los ingresos reales y los ingresos predichos.

```python
edad_nueva = [[30]]
prediccion = modelo.predict(edad_nueva)
print(prediccion)
```

```python
#podemos evaluar el modelo
from sklearn.metrics import r2_score, mean_squared_error
y_pred = modelo.predict(X) #son os ingresos que el modelo cree segun la edad
#ingresos = β0 + β1 · edad
#β0 → intercepto
#β1 → coeficiente de edad

r2 = r2_score(y, y_pred)
#0.3 - 0.5 aceptable, pero siempre es mejor tener un modelo 0.6 - 0.8
#o muy bueno > 0.8
mse = mean_squared_error(y, y_pred)
rmse = np-sqrt(mse)
print(r2)
#regresion logistica se suele usar -> Accuracy,Precicion,F1 Score, Recall,Roc Auc..
```
