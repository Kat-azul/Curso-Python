#Función lambda
Una función lambda es una función acprtada para escribir una función de esta forma
nombreFunción = lambda parámetros: acción
La diferencia entre la dunción lambda y funciones normales es que suelen ser un poco más limitadas

```python
sumar = lambda x,y : x+y
print(sumar(3,10))
```

```python
descuentoLambda = lambda desc, precio: ((precio* (1 - desc)), (precio*desc))
nuevoPrecio= descuentoLambda(0.4, 500)
print("El precio a pagar es", nuevoPrecio[0], "el descuento fue de: " )
print(type(nuevoPrecio))
```

```python
def promediosYmayor(promedios):
  if not promedios: # Manejar el caso de lista vacía
      return 0, 0 # O se podría lanzar un error o retornar None, None
  promedio = 0
  for x in promedios:
    promedio += x
  promedio = promedio / len(promedios)
  return promedio, max(promedios)

promedios = []
continuar = ""

while True:
  if len(promedios) > 0:
    continuar = ""
    while continuar not in ("si", "no"):
      continuar = input("¿Desea ingresar otra calificación? (si/no): ").lower()
    if continuar == "no":
      break

  while True:
    try:
      calif = int(input("Indique la calificación (0-100): "))
      if calif < 0 or calif > 100:
        print("Ingrese un valor válido (entre 0 y 100).")
      else:
        promedios.append(calif)
        break # Sale del bucle interior después de una entrada válida
    except ValueError:
      print("Ingrese un valor numérico.")

if promedios: # Solo calcular y imprimir si se ingresaron calificaciones
  prom, mayor = promediosYmayor(promedios)
  print("\nResultados:")
  print("Promedio:", prom)
  print("Mayor:", mayor)
else:
  print("No se ingresaron calificaciones.")
```

```python
#Calculadora = suma, resta, producto, división y potencia
def suma(a,b):
 return a+b
def resta(a,b):
 return a - b
def producto(a,b):
 return a*b
def division(a,b):
 if b==0:
  return "Error: no se puede dividir entre 0"
 return a/b
def potencia(a,b):
  return a**b

def menu():
    print("----------MI CALCULADORA--------")
    print("1. Suma")
    print("2. Resta")
    print("3. Producto")
    print("4. División")
    print("5. Potencia")
    print("6. Salir")

while True:
    menu()
    try:
        op = int(input("Seleccione una opción: "))
    except ValueError:
        print("Ingrese un número válido")
        continue

    if op == 6:
        print("SALIENDO...")
        break
    if op < 1 or op > 6:
        print("Opción inválida")
        continue

    try:
        a= float(input("Ingrese el primer número: "))
        b= float(input("Ingrese el segundo número: "))
    except ValueError:
        print("Error ingrese un número válido: ")
        continue

    if op == 1:
        print("Resultado:", suma(a, b))
    elif op == 2:
        print("Resultado:", resta(a, b))
    elif op == 3:
        print("Resultado:", producto(a, b))
    elif op == 4:
        resultado_division = division(a, b)
        if isinstance(resultado_division, str):
            print(resultado_division)
        else:
            print("Resultado:", resultado_division)
    elif op == 5:
        print("Resultado:", potencia(a, b))
```
